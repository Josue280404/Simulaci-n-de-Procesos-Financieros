```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

n_sim =5000
n_steps =200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()
```

```python
import numpy as np

np.random.seed(0)

n_sim = 2000
times = [10, 50, 100, 200, 400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1, 1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var{np.var(finals):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_pasos = 200
n_caminatas = 200

pasos = np.random.choice([-1, 1], size=(n_pasos, n_caminatas))

# Calculamos la suma acumulada por columna (axis=0)
# Sumamos el valor inicial de 100
caminatas = 100 + np.cumsum(pasos, axis=0)

plt.figure(figsize=(12, 6))
plt.plot(caminatas)
plt.plot(np.mean(caminatas, axis=1), color='black', linewidth=2, label='Promedio de las 200')

plt.title(f"Simulacin de {n_caminatas} Caminatas Aleatorias")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.axhline(100, color='black', linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
