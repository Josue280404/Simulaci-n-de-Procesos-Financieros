```python
def montecarlo_grafica(N=10000):
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)
    
    dentro = x**2 + y**2 <= 1
    
    # Grfica
    plt.figure(figsize=(6,6))
    plt.scatter(x[dentro], y[dentro], s=5, color='blue')
    plt.scatter(x[~dentro], y[~dentro], s=5, color='red')
    
    # Crculo
    theta = np.linspace(0, 2*np.pi, 300)
    plt.plot(np.cos(theta), np.sin(theta), color='black', linewidth=2)
    
    plt.axhline(0)
    plt.axvline(0)
    plt.gca().set_aspect('equal')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.legend()
    plt.title("Simulacin Monte Carlo")
    plt.show()

montecarlo_grafica(10000)
```
